### Identificando valores nulos en la columna 'Valor'

```python
# Identificando los valores nulos en la columna 'Valor'
datos['Valor'].isnull()# Este código creará una condición para crear un DataFrame con los valores nulos(NaN)
```

```python
#Creamos un DataFrame con los valores nulos de la columna 'Valor'
datos[datos['Valor'].isnull()]
```

```python
#Eliminaremos los valores nulos de la columna 'Valor' ya que no podemos inventar cualquier valor a estos inmuebles pues podría falsear nuestras conclusiones
A = datos.shape[0]# Total de filas antes de la eliminación de valores nulos
datos.dropna(subset=['Valor'],inplace=True)
B = datos.shape[0]# Total de filas despues de la eliminación de valores nulos
A-B# total filas eliminadas
```


###Identificando valores nulos en la columna 'Mantenimiento'

```python
datos
```

## Identificando y excluyendo Outliers

```python
datos.head(10)
```

```python
#Boxplot es un gráfico estadístico que nos permite visualizar nuestros datos
# corta nuestros datos en algunos cuartiles(Q1,Q2,Q3,) e identifica cuáles puntos o cuáles datos
# o cuáles líneas de nuestro banco están fuera de estos cortes
plt.rc('figure',figsize = (14,8))
datos.boxplot(['Valor'])
# En nuestro DataFrame,en la columna 'Valor', el boxplot ha creado índices, los rótulos de 0, 1, 2, 3 y 4,
# pero como está con un logaritmo al 6, imaginen que esto es 1.000.000, 2.000.000, 3.000.000, 4.000.000.
# Entonces, la mayoría de los datos está entre 0 y 1, es decir, entre 500.000,
# el valor está contenido prácticamente en torno de los 500.000 dólares
```

```python
# Vamos a visualizar los datos cuyo valor de alquiler estan fuera de los 500000 dólares
datos[datos['Valor'] >= 500000]
```

```python
#Entonces estos son nuestros outliers, tenemos que eliminarlos. Para eliminarlos, vamos a crear nuestros cuartiles
valor = datos['Valor']
Q1 = valor.quantile(.25) #quartil 1 = 25% de los datos
Q3 = valor.quantile(.75) #quartil 3 = 75% de los datos
IIQ = Q3 - Q1 # IIQ = Intervalo Inter Quartil
limite_inferior =  Q1 - 1.5*IIQ
limite_superior =  Q3 + 1.5*IIQ
```

## Identificando y excluyendo Outliers(continuación)


```python
# En el ejercicio anterior generamos nuestro box-plot y nos había indicado que teníamos valores que estaban
# muy por encima de la concentración de los datos y nos dedicamos a excluirlos, pero hay una columna que no
# llevamos en consideración, que es la columna tipo, algunos valores dependiendo del tipo serán más caros que otros.
# Por ejemplo una casa en condominio, el alquiler es mucho más caro que el alquiler de una habitación.
# Lo que vamos a hacer es generar los límites pero dependiendo del tipo de local
```

```python
datos.boxplot(['Valor'], by = ['Tipo'])
```

```python
# Vamos a crear los límites según el tipo de local que estemos considerando
grupo_tipo = datos.groupby('Tipo')# Crea un DataFrame que ha guardado todas las columnas agrupadas por la columna tipo.
type(grupo_tipo)
```

```python
# Lo que queremos es que únicamente almacene la información del valor así que vamos a colocar un parámetro más
# e indicarle que solo guarde la columna valor
grupo_tipo = datos.groupby('Tipo')['Valor']
type(grupo_tipo)# Ahora se generó una Series ya que solo muestra la columna 'Valor'
```

```python
# El atributo groups nos crea un diccionario, considerando la llave, cada tipo de local, y como valores
# las posiciones donde se encuentran nuestros locales.
grupo_tipo.groups
```

```python
# Usamos el código para crear las medidas para nuestro nuevo objeto, grupo_tipo.
Q1 = grupo_tipo.quantile(.25).round(2) #quartil 1 = 25% de los datos
Q3 = grupo_tipo.quantile(.75).round(2) #quartil 3 = 75% de los datos
IIQ = Q3 - Q1 # IIQ = Intervalo Inter Quartil
limite_inferior =  Q1 - 1.5*IIQ
limite_superior =  Q3 + 1.5*IIQ
```

```python
# Hemos creado un límite inferior y superior para cada tipo de local
limite_inferior
```

```python
limite_superior
```

```python
# Vamos a generar un DataFrame excluyendo los outliers según el tipo de cada local.
datos_new = pd.DataFrame() #Creamos un DataFrame en blanco
# vamos a recorrer cada tipo de mi objeto 'grupo_tipo', y vamos a llamar a su propiedad groups
# y a sus posiciones, keys.
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo #barremos la columna tipo del DataFrame original y preguntamos:"¿Esta fila, la columna tipo de esta fila, es la misma que está entrando en este for?" la respuesta sera un booleano
  eh_dentro_limite =(datos['Valor']>= limite_inferior[tipo]) & (datos['Valor']<= limite_superior[tipo])#la otra variable booleana nos va a preguntar si está dentro de los límites, si ambas condiciones se cumplen, esto resultará en un valor true
  seleccion = eh_tipo & eh_dentro_limite #para que la selección al final sea verdadera, ambas deben cumplir tanto que sean del mismo tipo y que esté dentro de los límites de ese tipo.
  datos_seleccion = datos[seleccion] #Creamos un DataFrame 'datos_selección' con el pedazo que cumpla de mis datos originales, que cumpla la selección que acabamos de crear
  datos_new = pd.concat([datos_new, datos_seleccion]) #Usamos el método concat()entre sus paréntesis tenemos que colocar el DataFrame donde vamos a guardar los datos, que es 'datos_new', y el DataFrame que voy a adicionar los nuevos valores, en este caso 'datos_selección'
```
